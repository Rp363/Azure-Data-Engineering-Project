### DimUser

#### Basic Loading(batch processing)

```python
df = spark.read.format('parquet')\
    .load('abfss://bronze@storageaccrojit.dfs.core.windows.net/DimUser')
```

```python
df.display()
```

#### Autoloader

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

df_user = spark.readStream.format('cloudFiles')\
          .option('cloudFiles.format','parquet')\
          .option('cloudFiles.schemaLocation', 'abfss://silver@storageaccrojit.dfs.core.windows.net/DimUserSchema/checkpoint')\
          .load('abfss://bronze@storageaccrojit.dfs.core.windows.net/DimUser')
```

```python
df_user.display()
```

```python
df_user = df_user.withColumn('user_name', upper(col('user_name')))
```

#### Creating the functions on util path so to use that we have to provide databricks a path which works as root path for this project to call or import that functions 

```python
import os
import sys

projectpath = os.path.join(os.getcwd(), '..', '..')
sys.path.append(projectpath)

from Utils.transformations import reuseable
```

#### Using our Utils functions

```python
df_user_obj = reuseable()
```

```python
df_user = df_user_obj.dropColumns(df_user, ['_rescued_data'])
df_user.display()
```

#### Back to normal

```python
df_user = df_user.dropDuplicates(["user_id"])
```

```python
df_user.writeStream.format('delta')\
       .outputMode('Append')\
       .option('checkpointLocation', 'abfss://silver@storageaccrojit.dfs.core.windows.net/DimUserSchema/checkpoint')\
       .trigger(once=True)\
       .option('path', 'abfss://silver@storageaccrojit.dfs.core.windows.net/DimUserSchema/Data')\
       .toTable('spotify_catalog.silver.DimUser')
```

### DimTrack

```python
df_track = spark.readStream.format('cloudFiles')\
                .option('cloudFiles.format', 'parquet')\
                .option('cloudFiles.schemaLocation', 'abfss://silver@storageaccrojit.dfs.core.windows.net/DimTrackSchema/checkpoint')\
                .load('abfss://bronze@storageaccrojit.dfs.core.windows.net/DimTrack')
display(df_track)
```

```python
df_track = df_track.withColumn('duration_flag', when(col('duration_sec')<=150, 'short').when(col('duration_sec')<=300, 'medium').otherwise('long'))

df_track = df_track.withColumn('track_name', regexp_replace(col("track_name"), '-', ' '))

display(df_track)
```

```python
df_track = reuseable().dropColumns(df_track, ['_rescued_data'])
```

```python
df_track.writeStream.format('delta')\
        .outputMode('Append')\
        .option('checkpointLocation', 'abfss://silver@storageaccrojit.dfs.core.windows.net/DimTrackSchema/checkpoint')\
        .trigger(once=True)\
        .option('path', 'abfss://silver@storageaccrojit.dfs.core.windows.net/DimTrackSchema/Data')\
        .toTable('spotify_catalog.silver.DimTrack')
```
